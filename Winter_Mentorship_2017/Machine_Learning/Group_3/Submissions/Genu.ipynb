```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
data = pd.read_csv('train.csv')
x1 = data['x1']
y = data['y']
m = np.size(x1)
x = np.ones(shape=(m,1))
x[:,0] = x1
alpha = 0.01
theta = np.ones(shape=(1,1))
def cost(x,yo,theta):
    pre = x.dot(theta)
    sqError = ( pre - yo )**2
    J = (1.0/(2*m))*sqError
    return J
def gradient(x,y,theta,alpha):
    J_his = np.zeros(shape=(m,1))
    for i in range(m):
        pre = x.dot(theta)
        error = ( pre - y[i] )*x[i]
        theta[0][0] = theta[0][0] - alpha * (1.0/m) * error.sum()
        J_his[i,0] = cost(x,y[i],theta).sum()
    return J_his , theta

theta[0][0]=1
l = gradient(x,y,theta,alpha)
c = l[1]
m = l[0][int(c)]
print('y =',m,'* x +',c)
plt.plot(x,m*x+c,'r')
        
data.plot()
plt.show()
```
